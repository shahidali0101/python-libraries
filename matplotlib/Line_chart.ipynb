# _______________________*Line_chart*.

```python
import matplotlib.pyplot as plt
```

## *Charts customization*

```python
# creating line chart 

x=[1,2,3,4,5]
y=[1,4,9,16,25]

plt.plot(x,y)
plt.xlabel('X - Axis')
plt.ylabel('Y- Axis')
plt.title("Basic Line Plot")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

plt.plot(x,y,color='red',linestyle='solid',linewidth=2,marker='o',markersize=5,mfc='green')
plt.xlabel('X - Axis')
plt.ylabel('Y- Axis')
plt.title("Basic Line Plot")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

plt.plot(x,y,color='green',linestyle='solid',linewidth=2,marker='o',markersize=5,mfc='red')
plt.grid(True)
plt.xlabel('X - Axis')
plt.ylabel('Y- Axis')
plt.title("Basic Line Plot")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

plt.plot(x,y,color='red',linestyle='solid',linewidth=2,marker='o',markersize=5,mfc='green')
plt.xlabel('X - Axis')
plt.ylabel('Y- Axis')
plt.title("Basic Line Plot",color='red')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

plt.plot(x,y,color='blue',linestyle='solid',linewidth=2,marker='o',markersize=5,mfc='red')
plt.xlabel('X - Axis')
plt.ylabel('Y- Axis')
plt.title("Basic Line Plot",color='black',family='Times New Roman',size=20)
plt.show()
```

###       *multiplots*

```python
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,6]
plt.figure(figsize=(9,5))

plt.subplot(1,2,1)  # 1 one 2 column 2 chart
plt.plot(x,y1,color='green')
plt.title('PLOT-1',color='green')

plt.subplot(1,2,2)  # 1 one 2 column 
plt.plot(x,y2,color='red')
plt.title('PLOT-2',color='red')
```

```python
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,6]
plt.figure(figsize=(9,5))

plt.subplot(2,2,1)  # 1 one 2 column 2 chart
plt.plot(x,y1,color='green')
plt.title('PLOT-1',color='green')

plt.subplot(2,2,2)  # 1 one 2 column 
plt.plot(x,y2,color='red')
plt.title('PLOT-2',color='red')

plt.subplot(2,2,3)  # 1 one 2 column 2 chart
plt.plot(x,y1,color='blue')
plt.title('PLOT-3',color='green')
```

```python
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,6]
plt.figure(figsize=(9,5))

plt.subplot(2,2,1) 
plt.plot(x,y1,color='green')
plt.title('PLOT-1',color='green')

plt.subplot(2,2,2) 
plt.plot(x,y2,color='red')
plt.title('PLOT-2',color='red')

plt.subplot(3,1,3)  
plt.plot(x,y1,color='blue')
plt.title('PLOT-3',color='green')

```

```python
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,6]
plt.figure(figsize=(9,8))

plt.subplot(2,2,1) 
plt.plot(x,y1,color='green')
plt.title('PLOT-1',color='green')

plt.subplot(2,2,2) 
plt.plot(x,y2,color='red')
plt.title('PLOT-2',color='red')

plt.subplot(2,2,3)  
plt.plot(x,y1,color='blue')
plt.title('PLOT-3',color='green')

plt.subplot(2,2,4) 
plt.plot(y1,x,color='black')
plt.title('PLOT-4',color='black')
```

```python
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,6]

plt.figure(figsize=(9,8))



plt.subplot(2,2,1) 
plt.plot(x,y1,color='green')
plt.title('PLOT-1',color='green')
plt.grid(True)

plt.subplot(2,2,2) 
plt.plot(x,y2,color='red')
plt.title('PLOT-2',color='red')
plt.grid(True)

plt.subplot(2,2,3)  
plt.plot(x,y1,color='blue')
plt.title('PLOT-3',color='green')
plt.grid(True)

plt.subplot(2,2,4) 
plt.plot(y1,x,color='black',marker="o",mfc='blue')
plt.title('PLOT-4',color='black')
```
