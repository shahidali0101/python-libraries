```python
link1="C:\\Users\\Shahid ali\\Desktop\\brillica\\Nike Data.xlsx"
```

```python
nk=pd.read_excel(link1)
nk.head(1)
```

```python
# Create a column chart to show the sum of unit sold by each state.

nk.groupby("State")["Units Sold"].sum().reset_index()
x=nk["State"]
y=nk["Units Sold"]

plt.figure(figsize=(20,5))
plt.bar(x,y,color="red")
plt.xlabel("State")
plt.ylabel("Sum of units sold")
plt.xticks(rotation=60)
plt.title("Sum of units sold by state",color="green")
plt.show() 
```

```python
#show the maximum total sales by each city in bar chart
c=nk.groupby("City")["Total Sales"].max().reset_index()

x=c["City"]
y=c["Total Sales"]

plt.figure(figsize=(20,5))
plt.bar(x,y,color="blue")
plt.xlabel("city")
plt.ylabel("tatal sales")
plt.xticks(rotation=60)
plt.title("Sum of units sold by state",color="green")
plt.show() 
```

```python
# Create a pie chart to show the region by sum of unit price 
p=nk.groupby("Region")["Price per Unit"].sum().reset_index()
p
y=p["Price per Unit"]
x=p["Region"]
plt.pie(y,labels=x,autopct="%0.1f%%")
plt.show()
```
