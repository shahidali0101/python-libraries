```python
import matplotlib.pyplot as plt
```

### Matplotlib (the Matplotlib library) supports many types of charts for data visualization
1. line chart
2. Graph
3. Histogram
4. Pie chart
5. Scatter plot
6. box plot 


## 1. *line chart*
*Used for trend over time*

```python
x = [1,2,3,4]
y = [10,20,25,30]

plt.plot(x, y)
plt.show()
```

## 2. *Bar chart*
*it is used for compare Categories*

```python
plt.bar(['A','B','C'], [10,20,15])
plt.show()
```

## *3. Histogram*
*Shows distribution of data*

```python
plt.hist([12, 15, 14, 10, 18, 20, 22, 19, 17, 16, 15, 14, 13, 12, 11, 10])
plt.show()
```

##  *4. Pie chart*
*Pie charts are best for showing percentage or proportion distribution of data.*

```python
plt.pie([40,30,30], labels=['A','B','C'])
plt.show()
```

## *5. Scatter Plot*
- Shows relationship between variables
  - Example: height vs weight

```python
plt.scatter([1,2,3], [4,5,6])
plt.show()
```

## *6. Area chart*
- Similar to line chart but filled area

```python
plt.fill_between([1,2,3], [2,4,6])
plt.show()
```

## *7. Box plot*
*Shows spread, median, outliers*    
*A box plot is used to summarize the distribution of data and quickly understand how values are spread.*


```python
plt.boxplot([1,2,3,4,5,10])
plt.show()
```

## *8.Heat Map*
*A heatmap is used to visualize data intensity using colors. Instead of reading numbers in a table, you can quickly understand patterns through color difference*

```python
data = [[1,2,3],
        [4,5,6],
        [7,8,9]]

plt.imshow(data, cmap='coolwarm')
plt.colorbar()
plt.show()
```
